```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Pie chart
pie_df=data.groupby('company', axis=0).count()
pie_df['account_type'].plot(kind='pie',
                    figsize=(8,8),
                    autopct='%1.1f%%',
                    explode=(0.3,0,0),
                    colors=['green','red','blue'],
                    labels=['mm_account_telco'])
plt.legend(labels==pie_df.index, loc='upper left')
plt.title('Division of mobile money market')
plt.show()
```

```python
labels=['Company A','Company B','Company C']
values=[1,2,3]
colors=['green','red','blue']
explode=[0.3,0,0]
plt.title=('Division of Mobile Money Market')
plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```
