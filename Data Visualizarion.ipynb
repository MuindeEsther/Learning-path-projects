```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels =['Nokia','Samsung','Apple','Lumia']
values =[10,30,45,15]
colors =['yellow','green','red','blue']
plt.pie(values, labels=labels, colors=colors)
plt.axis('equal')
```

```python
# we can add complexity by draw it with a slice extracted from it i.e Nokia
labels =['Nokia','Samsung','Apple','Lumia']
values =[10,30,45,15]
colors =['yellow','green','red','blue']
explode =[0.3,0,0,0]
plt.title =('A Pie Chart')
plt.pie(values, labels=labels, colors=colors, explode=explode, startangle=180)
plt.axis('equal')
```

```python
# To introduce the aspect of percentage, we use autopct and to make it more appealing we use shadow kwarg
labels =['Nokia','Samsung','Apple','Lumia']
values =[10,30,45,15]
colors =['yellow','green','red','blue']
explode =[0.3,0,0,0]
plt.title =('A Pie Chart')
plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%1.1f%%', startangle=180)
plt.axis('equal')
```

Pie charts with Pandas DataFrame.


```python
data = {'series1':[1,3,4,3,5],
        'series2':[2,4,5,2,4],
        'series3':[3,2,3,1,3]}
df = pd.Dataframe(data)
df['series1'].plot(kind='pie', figsize=(6,6))
```
